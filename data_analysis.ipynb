```python
import pandas as pd
import seaborn as sns
import matplotlib as plt
import matplotlib.pyplot as mplt
```

```python
template_df = pd.read_csv(
    './data/ML-CUP23-Template_example.csv',
    index_col=0,
    comment='#',
    header=None,
    names=['output_' + str(i+1) for i in range(3)]
)
training_df = pd.read_csv(
    './data/ML-CUP23-TR.csv',
    index_col=0,
    comment='#',
    header=None,
    names=['feature_' + str(i+1) for i in range(10)] + ['target_' + str(i+1) for i in range(3)]
)
test_df = pd.read_csv(
    './data/ML-CUP23-TS.csv',
    index_col=0,
    comment='#',
    header=None,
    names=['feature_' + str(i+1) for i in range(10)]
)

features_columns = ['feature_' + str(i+1) for i in range(10)]
targhet_columns = ['target_' + str(i+1) for i in range(3)]
```

```python
fig, ax = mplt.subplots(figsize=(15, 5))
sns.violinplot(data=test_df,ax=ax)
ax.yaxis.grid()
```

```python
fig, ax = mplt.subplots(figsize=(15, 5))
sns.violinplot(data=training_df[features_columns],ax=ax)
ax.yaxis.grid()
```

```python
fig, ax = mplt.subplots(figsize=(15, 5))
sns.violinplot(data=training_df[targhet_columns],ax=ax)
ax.yaxis.grid()
```

```python
fig, ax = mplt.subplots(figsize=(15, 15))
pd.plotting.scatter_matrix(training_df[features_columns], ax=ax);
```

```python
fig, ax = mplt.subplots(figsize=(15, 12))
sns.heatmap(data=training_df[features_columns].corr(), vmin=-1, vmax=1, annot=True, ax=ax)
```
