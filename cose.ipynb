```python
import pandas as pd
import seaborn as sns
import matplotlib as plt
import matplotlib.pyplot as mplt
```

```python
template_df = pd.read_csv(
    './data/ML-CUP23-Template_example.csv',
    index_col=0,
    comment='#',
    header=None,
    names=['output_' + str(i+1) for i in range(3)]
)
training_df = pd.read_csv(
    './data/ML-CUP23-TR.csv',
    index_col=0,
    comment='#',
    header=None,
    names=['feature_' + str(i+1) for i in range(10)] + ['target_' + str(i+1) for i in range(3)]
)
test_df = pd.read_csv(
    './data/ML-CUP23-TS.csv',
    index_col=0,
    comment='#',
    header=None,
    names=['feature_' + str(i+1) for i in range(10)] + ['target_' + str(i+1) for i in range(3)]
)
```
